```python
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# ---------------- CONFIGURATION ---------------- #
TOP_K_DEFAULT = 20
MAX_DECODE_STEPS_DEFAULT = 20
PROBABILITY_COLORMAP = "viridis"
ATTENTION_COLORMAP = "YlGnBu"

# ---------------- UTILITY FUNCTIONS ---------------- #

def print_vectors(tensor, title, limit=5):
    """Prints a tensor's values in a limited, formatted way."""
    print(f"\n--- {title} ---")
    if tensor.dim() == 2:
        print(f"Shape: {tensor.shape}")
        for i in range(min(limit, tensor.shape[0])):
            row = tensor[i, :limit].tolist()
            print(f"Token {i}: {[f'{x:.4f}' for x in row]} ...")
    else:
        print(f"Shape: {tensor.shape}")
        row = tensor[:limit].tolist()
        print(f"Vector: {[f'{x:.4f}' for x in row]} ...")

def visualize_probabilities_plot(probs, id_to_token, top_k=TOP_K_DEFAULT, cmap=PROBABILITY_COLORMAP):
    """
    Visualizes the top-k token probabilities using a graphical bar chart.
    """
    if probs.dim() != 1:
        raise ValueError("Expected probability tensor of shape [vocab_size]")

    top_k = min(top_k, probs.shape[0])
    top_probs, top_indices = torch.topk(probs, top_k)

    tokens = [id_to_token.get(idx.item(), f"<id:{idx.item()}>") for idx in top_indices]

    plt.figure(figsize=(10, 6))
    sns.barplot(x=tokens, y=top_probs.tolist(), palette=cmap)

    plt.title(f"Top {top_k} Predicted Tokens and Their Probabilities")
    plt.xlabel("Predicted Token")
    plt.ylabel("Probability")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

def plot_positional_encoding(seq_len, d_model):
    """Creates a plot of sinusoidal positional encodings."""
    if seq_len <= 0 or d_model <= 0:
        raise ValueError("Invalid sequence length or model dimension")

    position = torch.arange(seq_len).unsqueeze(1)
    div_term = torch.exp(torch.arange(0, d_model, 2) * (-np.log(10000.0) / d_model))

    pe = torch.zeros(seq_len, d_model)
    pe[:, 0::2] = torch.sin(position * div_term)
    pe[:, 1::2] = torch.cos(position * div_term)

    plt.figure(figsize=(10, 6))
    for i in range(min(5, d_model)):
        plt.plot(pe[:, i].numpy(), label=f"dim {i}")

    plt.title("Sinusoidal Positional Encodings")
    plt.xlabel("Position")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

def plot_attention_heatmap(
    attention_matrix,
    x_labels,
    y_labels,
    title,
    cmap=ATTENTION_COLORMAP,
    max_annot_cells=25
):
    """
    Visualizes an attention matrix with adaptive scaling to prevent blur.
    """

    attention_matrix = attention_matrix.detach().cpu()
    if attention_matrix.dim() != 2:
        raise ValueError("Attention matrix must be 2D")

    rows, cols = attention_matrix.shape

    # ---- Adaptive Figure Size ---- #
    cell_size = 0.6
    fig_width = max(6, cols * cell_size)
    fig_height = max(2, rows * cell_size)

    plt.figure(figsize=(fig_width, fig_height), dpi=120)

    # ---- Conditional Annotation ---- #
    annotate = (rows * cols) <= max_annot_cells

    sns.heatmap(
        attention_matrix.detach().cpu().numpy(),
        xticklabels=x_labels,
        yticklabels=y_labels,
        cmap=cmap,
        linewidths=0.3,
        linecolor="gray",
        annot=annotate,
        fmt=".2f" if annotate else "",
        annot_kws={"size": 9} if annotate else None,
        cbar=True
    )

    plt.title(title, fontsize=12)
    plt.xlabel("Keys (Input Tokens)")
    plt.ylabel("Queries (Output Tokens)")

    plt.xticks(rotation=45, ha="right", fontsize=9)
    plt.yticks(rotation=0, fontsize=9)

    plt.tight_layout()
    plt.show()

# ---------------- MAIN PIPELINE ---------------- #

def run_translation(
    top_k=TOP_K_DEFAULT,
    max_decode_steps=MAX_DECODE_STEPS_DEFAULT
):
    # ---- Model Loading with Safety ---- #
    model_name = "Helsinki-NLP/opus-mt-en-de"
    try:
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    except Exception as e:
        raise RuntimeError(f"Failed to load model or tokenizer: {e}")

    # ---- Input Validation ---- #
    input_sentence = input("Enter a sentence to translate: ")
    if not isinstance(input_sentence, str) or not input_sentence.strip():
        raise ValueError("Input sentence must be a non-empty string")

    enc_inputs = tokenizer(input_sentence, return_tensors="pt")
    if enc_inputs.input_ids.dim() != 2:
        raise ValueError("Unexpected encoder input shape")

    enc_tokens = tokenizer.convert_ids_to_tokens(enc_inputs.input_ids[0])

    print(f'\nInput Sentence: "{input_sentence}"')
    print("Tokenized input IDs:", enc_inputs.input_ids)
    print("Tokenized input tokens:", enc_tokens)

    # ---- Positional Encoding Visualization ---- #
    plot_positional_encoding(enc_inputs.input_ids.shape[1], model.config.d_model)

    # ---- Embedding Inspection ---- #
    embeddings = model.get_input_embeddings()
    input_embeddings = embeddings(enc_inputs.input_ids)

    if input_embeddings.dim() != 3:
        raise ValueError("Unexpected embedding tensor shape")

    print_vectors(input_embeddings[0], "Input Embeddings (First 5 dimensions)")

    # ---- Vocabulary Mapping ---- #
    vocab = tokenizer.get_vocab()
    id_to_token = {v: k for k, v in vocab.items()}

    # ---- Autoregressive Decoding ---- #
    generated_ids = torch.tensor([[model.config.decoder_start_token_id]])
    dec_tokens = [tokenizer.bos_token]

    print("\n--- Autoregressive Decoding ---")

    with torch.no_grad():
        for step in range(max_decode_steps):
            outputs = model(
                input_ids=enc_inputs.input_ids,
                decoder_input_ids=generated_ids,
                output_attentions=True
            )

            logits_last = outputs.logits[:, -1, :].squeeze(0)
            probabilities = F.softmax(logits_last, dim=-1)

            next_token_id = torch.argmax(logits_last).view(1, 1)
            predicted_token = tokenizer.decode(next_token_id[0], skip_special_tokens=True)

            print(f"\nStep {step + 1}: Predicted token is '{predicted_token}'")
            visualize_probabilities_plot(probabilities, id_to_token, top_k=top_k)

            generated_ids = torch.cat([generated_ids, next_token_id], dim=-1)
            dec_tokens.append(predicted_token)

            if next_token_id.item() == tokenizer.eos_token_id:
                break

    final_translation = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    print("\nFinal Translation:", final_translation)

    # ---- Attention Visualization ---- #
    outputs_final = model(
        input_ids=enc_inputs.input_ids,
        decoder_input_ids=generated_ids,
        output_attentions=True
    )

    cross_attn = outputs_final.cross_attentions[-1][0, -1]
    plot_attention_heatmap(
        cross_attn,
        enc_tokens,
        [dec_tokens[-1]],
        "Cross-Attention (Last Token)"
    )

    self_attn = outputs_final.decoder_attentions[-1][0, -1]
    plot_attention_heatmap(
        self_attn,
        dec_tokens,
        [dec_tokens[-1]],
        "Masked Self-Attention (Last Token)"
    )

# ---------------- ENTRY POINT ---------------- #

if __name__ == "__main__":
    run_translation()
```
